## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
orders.shape
```

```python
#Sub Problem 1: How to aggregate the amount_spent for unique customers?
# First Groupby customer and Country and then aggregate their amount_spent
aggregation_spent = orders.groupby(["CustomerID", "Country"]).agg({"amount_spent": sum}).reset_index()
```

```python
aggregation_spent
```

```python
#This is the value at 0.95 quantile
aggregation_spent.amount_spent.quantile(0.95) 
```

```python
#This filter gives me the VIP customers whose aggregate amount_spent is above the 0.95 quantile 
# I will assign this to a variable called vip_customers which only contain vip customers, their country and aggregation spent. 

vip_customers = aggregation_spent[aggregation_spent['amount_spent'] > aggregation_spent.amount_spent.quantile(0.95)]
vip_customers
```

```python
#This is the value at 0.75 quantile
aggregation_spent.amount_spent.quantile(0.75)
```

```python
#This filter will give me the preferred customers in between 75th quantile and 95th quantile
# Will create another variable called preferred_customers and will store each preferred customer, their country and aggregated
#amount spent. 
preferred_customers = aggregation_spent[(aggregation_spent['amount_spent'] > aggregation_spent.amount_spent.quantile(0.75))
                                       &(aggregation_spent['amount_spent'] < aggregation_spent.amount_spent.quantile(0.95))]
```

```python
preferred_customers
```

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
#Easy way to look at it woul be just to see the value counts from the vip_customers variable


vip_customers['Country'].value_counts()
#With this I see majority come from the United Kingdom
```

```python
#Another way to do it would be with the .groupby
vip_customers.groupby(['Country'])['Country'].count().sort_values(ascending=False)
```

**Majority of VIP customers come from UK.** 

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
# The straightforward way for me to do it, would be to combine both the VIP and Preferred Customers variables, and then do the couns
# We concat the two dataframes on axis = 0

vip_preferred_customers = pd.concat([vip_customers, preferred_customers], axis = 0)
```

```python
vip_preferred_customers
```

```python
vip_preferred_customers.groupby(['Country'])['Country'].count().sort_values(ascending=False)
```

**Majority of VIP + Preferred customers come from UK.**
